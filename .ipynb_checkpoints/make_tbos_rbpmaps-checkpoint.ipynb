# RBP Maps 0.2 can be used as a reference for SE events
- This notebook verifies the changes made do not affect the last time dataframes were checked independently. 

```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import sys
import math
import glob
import qtools
sys.path.insert(0, '/home/bay001/projects/codebase/rbp-maps/maps/')
from density import Map
from density import ReadDensity
from density import normalization_functions
from density import RDPlotter
pd.set_option('max_colwidth', 1000)
pd.set_option('display.max_rows', 1500)
```

```python
wd = '/projects/ps-yeolab3/bay001/tbos/images/rbpmaps_3-6-2017/'
program = '/home/bay001/projects/codebase/rbp-maps/maps/plot_density.py'
manifest = '/projects/ps-yeolab3/bay001/tbos/input_norm/input_norm_latest/input_norm_manifest.tsv'
annotation_dir = '/projects/ps-yeolab3/bay001/tbos/annotations/exon_centric_as_miso/3-6-2017/'
manifest = pd.read_table(manifest)
```

```python
cmd = "python {}".format(program)
```

```python
rbp = 'CSTF1'
for col, row in manifest[manifest['RBP']==rbp].iterrows():
    pos_ip_bw = row['CLIP'].replace('.bam','.norm.pos.bw')
    neg_ip_bw = row['CLIP'].replace('.bam','.norm.neg.bw')
    pos_in_bw = row['INPUT'].replace('.bam','.norm.pos.bw')
    neg_in_bw = row['INPUT'].replace('.bam','.norm.neg.bw')
    
    output_file = os.path.join(wd,"{}-{}.svg".format(rbp, row['uID']))
    annotations = glob.glob(annotation_dir + "alt_cassettes_{}_over_*.csv".format(rbp.lower()))
    cmd = "python {} ".format(program)
    cmd = cmd + '--ip {} '.format(row['CLIP'])
    cmd = cmd + '--input {} '.format(row['INPUT'])
    cmd = cmd + '-a {} {} {} '.format(annotations[0], annotations[1], annotations[2])
    cmd = cmd + '-t {} {} {} '.format('miso','miso','miso')
    cmd = cmd + '-e se '
    cmd = cmd + '--exon_offset {} '.format(50)
    cmd = cmd + '--intron_offset {} '.format(400)
    cmd = cmd + '-o {}'.format(output_file)
    print(cmd)
    # qtools.Submitter(cmd, output_file, array=False, nodes=1, ppn=1, walltime='1:00:00', submit=True)
```

# make objects and run that way

```python
print(annotation_dir + "alt_cassettes_{}_over*included*.tsv".format(rbp.lower()))
```

```python
rbp = 'CSTF1'
f, (ax1, ax2, ax3, ax4) = plt.subplots(
    1, 4, sharey=True, figsize=(16,4)
)
axs = [ax1, ax2, ax3, ax4]

for col, row in manifest[manifest['RBP']==rbp].iterrows():
    output_file = os.path.join(wd,"{}-{}.svg".format(rbp, row['uID']))
    pos_ip_bw = row['CLIP'].replace('.bam','.norm.neg.bw')
    neg_ip_bw = row['CLIP'].replace('.bam','.norm.pos.bw')
    pos_in_bw = row['INPUT'].replace('.bam','.norm.neg.bw')
    neg_in_bw = row['INPUT'].replace('.bam','.norm.pos.bw')
    ip_bam = row['CLIP']
    inp_bam = row['INPUT']
    included = glob.glob(annotation_dir + "alt_cassettes_{}_over*included*.tsv".format(rbp.lower()))[0]
    excluded = glob.glob(annotation_dir + "alt_cassettes_{}_over*excluded*.tsv".format(rbp.lower()))[0]
    unchanged = glob.glob(annotation_dir + "alt_cassettes_{}_over*unchanged*.tsv".format(rbp.lower()))[0]
    annotations = {included:'miso', excluded:'miso', unchanged:'miso'}
    print(unchanged)
    ip_rd = ReadDensity.ReadDensity(pos=pos_ip_bw,neg=neg_ip_bw,bam=ip_bam)
    in_rd = ReadDensity.ReadDensity(pos=pos_in_bw,neg=neg_in_bw,bam=inp_bam)
    rbpmap = Map.SkippedExon(
        ip=ip_rd, inp=in_rd, output_filename=output_file,
        norm_function=normalization_functions.normalize_and_per_region_subtract,
        annotation=annotations,intron_offset=400,
    )
    rbpmap.create_matrices()
    print('creating matrices DONE')
    rbpmap.normalize_matrix()
    print('normalizing DONE')
    rbpmap.set_means_and_sems()
    print('setting mean and sems DONE')
    RDPlotter.plot_se(rbpmap.means, rbpmap.sems, axs)
plt.tight_layout(pad=8, w_pad=3, h_pad=5)
f.suptitle(rbp)
f.savefig('/home/bay001/{}.svg'.format(rbp))
```
