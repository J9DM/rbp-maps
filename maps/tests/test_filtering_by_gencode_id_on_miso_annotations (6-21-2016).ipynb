```python
import pandas as pd
import numpy as np
```

```python
se = pd.read_table('/home/bay001/projects/maps_20160420/permanent_data/miso_se_to_ensembl.tsv',names=['event','name'])
diffexp = ['ENSG00000140575.8','ENSG00000254673.1','ENSG00000167074.10','ENSG00000143479.11','ENSG00000119514.5','ENSG00000258465.2','ENSG00000163931.1']
```

```python
def ensembl_from_gencode(gencode_id):
    return gencode_id.split('.')[0]
```

```python
diffexp = [ensembl_from_gencode(f) for f in diffexp]
```

```python
se['gene'] = se['name'].str.split(',')
se = se.dropna()
def isin(row,lst):
    for g in row['gene']:
        if g in lst:
            return True
    return False
se['isinlist'] = se.apply(isin,args=[diffexp],axis=1)
se = se[se['isinlist']==True]
del se['isinlist']
del se['gene']
```
