```python
%matplotlib inline
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '/home/bay001/projects/codebase/rbp-maps/maps/')
import pandas as pd
import numpy as np
import glob
import os
from density import normalization_functions as norm
from plot import Plot
```

```python
wd = '/projects/ps-yeolab3/bay001/maps/current/se_nr/'
files = glob.glob(
    os.path.join(wd,'*.nr.feature.se.raw_density_matrix.csv')
)
```

```python
f_ast = []
for f in files:
    f = f.replace('.positive','*').replace('.negative','*')
    f_ast.append(f)
f_ast = list(set(f_ast))
f_ast[:2]
```

```python
outdir = '/projects/ps-yeolab3/bay001/maps/current/se_just_raw_density/'
err['region1'] = list(pd.Series(0, range(0,1400)))

for f in f_ast:
    negative = {}
    positive = {}
    n, p = sorted(glob.glob(f))
    if 'negative' not in n and 'positive' not in p:
        print('uhoh')
    else:
        print(
            os.path.basename(n).replace(
                '.negative.nr.feature.se.raw_density_matrix.csv','.raw_density.png'
            )
        )
        outfile = os.path.join(
            outdir,os.path.basename(n).replace(
                '.negative.nr.feature.se.raw_density_matrix.csv','.raw_density.png'
            )
        )
                               
        
        negative['region1'] = list(pd.read_table(n,sep=',',index_col=0).mean())
        positive['region1'] = list(pd.read_table(p,sep=',',index_col=0).mean())
        title = os.path.basename(n).replace(
                '.negative.nr.feature.se.raw_density_matrix.csv',''
            )
        Plot.plot_se(positive, negative, positive, err, err, title, outfile)
```
